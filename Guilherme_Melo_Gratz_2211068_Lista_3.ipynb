```python
import pandas as pd
import pickle as pkl
from google.colab import  drive
import random
```

```python
# Q1 & Q2

drive.mount('/drive')
diabetes = pd.read_csv('/drive/MyDrive/PussyRio/eng1111/pratica1-classificacao/diabetes/class_diabetes.csv')

pd.set_option('display.max_rows', 3)
display(diabetes)
```

```python
#Q3

diabetes["insu"].dropna() #no caso deste banco de dados, nao ha valor algum em coluna alguma com valor nulo, portanto nenhuma linha foi apagada.

pd.set_option('display.max_rows', 5)
display(diabetes)
```

```python
#Q4

mediaInsu = diabetes["insu"].mean()
print(mediaInsu)
```

```python
#Q5

pd.set_option('display.max_rows', 5)

lin1Excluir = diabetes[ abs(diabetes["insu"] - mediaInsu)/mediaInsu > 0.60 ].index
diabetes.drop(lin1Excluir, inplace = True)

novodf = diabetes

display(novodf)
```

```python
#Q6 

c1= novodf['insu']
c2= novodf['mass']
c3= novodf['pedi']
c4= novodf['age']
c5= novodf['class']

diabetes5Col = pd.concat([ c1, c2, c3, c4, c5 ], axis = 1)
display(diabetes5Col)
```

```python
#Q7 

colunaA = diabetes5Col['mass']
colunaB = diabetes5Col['age']
numRand = random.randint(1,10)/10

print('multiplicando por %.1f', numRand)

diabetes5Col.insert( 4, 'mass_age', ((colunaA - colunaB)*numRand))
pd.options.display.float_format = '{:.1f}'.format

display(diabetes5Col)
```

```python
#Q8

import numpy as np

diabetes5Col = diabetes5Col.reindex(np.random.permutation(diabetes5Col.index))

display(diabetes5Col)
```

```python
#Q9

#Ordenarei o dataframe pelo peso das pessoas que nele aparecem (do maior para o menor)

diabetes5Col = diabetes5Col.sort_values(by='mass', ascending=False)

display(diabetes5Col)
```

```python
#Q10

grpInsulinas = diabetes5Col.groupby('insu')[['insu']].count()
grpTestado = diabetes5Col.groupby('class')[['class']].count()

display(grpInsulinas)
display(grpTestado)
```

```python
#Q11

#tranformando "tested_positive" em 1, e "tested_negative" em 0 na propria coluna:

diabetes5Col["class"] = diabetes5Col["class"].map(
    {'tested_negative' : 0, 'tested_positive' : 1}
)

display(diabetes5Col)
display(diabetes5Col['class'])
```
